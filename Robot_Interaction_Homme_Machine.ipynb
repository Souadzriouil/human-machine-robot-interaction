```python
import os
import speech_recognition as sr  
import pyttsx3 
import gymnasium as gym  
import time
import keyboard  


engine = pyttsx3.init() 
recognizer = sr.Recognizer() 


mic_index = 2 


env = gym.make("MountainCar-v0", render_mode="human")


COMMANDS = {
    "avance": 2, 
    "recule": 0,  
    "stop": 1  
}


def speak(text):
    """Le robot parle."""
    engine.say(text)
    engine.runAndWait()


def recognize_speech():
    """Utilise uniquement Google Speech pour la reconnaissance vocale."""
    with sr.Microphone(device_index=mic_index) as source:
        print("🎤 Parlez maintenant...")
        recognizer.adjust_for_ambient_noise(source, duration=1)  
        audio = recognizer.listen(source)

    try:
        command = recognizer.recognize_google(audio, language="fr-FR")
        print(f"✅ Reconnu : {command}")
        return command.lower()
    except sr.UnknownValueError:
        print("❌ Je n’ai pas compris.")
    except sr.RequestError:
        print("⚠️ Erreur de connexion.")

    return None


def confirm_command(command):
    """Demande une confirmation avant d'exécuter l'action."""
    speak(f"As-tu dit {command}? Réponds par oui ou non.")
    confirmation = recognize_speech()

    return confirmation == "oui"


obs, _ = env.reset()

while True:
    speak("Donnez une commande : avance, recule ou stop")
    command = recognize_speech()

    if command is None:
        continue  

    
    if keyboard.is_pressed("q"):  
        speak("Le robot s'arrête.")
        env.close()
        break  

    if "stop" in command:
        speak("Le robot s'arrête.")
        env.close()
        break  

    action = COMMANDS.get(command, None)

    if action is None:
        speak("Commande non reconnue. Essayez : avance, recule ou stop.")
        continue  

    print(f"🟡 Action détectée : {action}")  

    if confirm_command(command):  
        for _ in range(20):  
            obs, _, done, _, _ = env.step(action)
            env.render()
            time.sleep(0.05)

            if done: 
                obs, _ = env.reset()

        speak(f"Le robot {command}.")
    else:
        speak("D'accord, essaie encore.")
```

```python
import cv2
import mediapipe as mp
import gymnasium as gym
import highway_env
import numpy as np
import time
import pygame
import os


pygame.init()


sound_path = r"D:\Downloads\472615__timothyd4y__vintage-race-car-passing-2.wav"


if os.path.exists(sound_path):
    acceleration_sound = pygame.mixer.Sound(sound_path)
    print("🔊 Fichier son chargé avec succès !")
else:
    acceleration_sound = None
    print(f"❌ Le fichier son est introuvable : {sound_path}")


mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.7, min_tracking_confidence=0.7)
mp_drawing = mp.solutions.drawing_utils


GESTURES = {
    "left": 3,  
    "right": 4,  
    "up": 2,  
    "down": 0  
}


LEVELS = [
    {"vehicles_count": 5, "lanes_count": 2, "goal": 200},
    {"vehicles_count": 10, "lanes_count": 3, "goal": 300},
    {"vehicles_count": 15, "lanes_count": 4, "goal": 400},
    {"vehicles_count": 20, "lanes_count": 5, "goal": 500}
]

current_level = 0

def setup_env(level):
    """Configure l'environnement en fonction du niveau."""
    global env
    
 
    if env is not None:
        env.close()

    env = gym.make("highway-v0", render_mode="human")
    env.unwrapped.configure({
        "manual_control": False,
        "vehicles_count": LEVELS[level]["vehicles_count"],
        "duration": 100,
        "policy_frequency": 5,
        "lanes_count": LEVELS[level]["lanes_count"],
        "initial_speed": 30,
        "offroad_terminal": False
    })
    env.reset()



env = None
setup_env(current_level)


score = 0
game_over = False  
start_time = time.time()
overtake_count = 0  


def stop(message):
    global cap, env, game_over
    print(message)

   
    final_frame = np.zeros((400, 800, 3), dtype=np.uint8)
    color = (0, 255, 0) if "VICTOIRE" in message else (0, 0, 255)
    cv2.putText(final_frame, message, (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2, cv2.LINE_AA)
    cv2.imshow("FIN DU JEU", final_frame)
    cv2.waitKey(2000)

   
    game_over = True  
    cap.release()
    env.close()
    cv2.destroyAllWindows()
    exit()

def detect_gesture(hand_landmarks):
    if hand_landmarks:
        index_tip = hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP]
        wrist = hand_landmarks.landmark[mp_hands.HandLandmark.WRIST]

        if index_tip.x < wrist.x - 0.1:  
            return "left"
        elif index_tip.x > wrist.x + 0.1:  
            return "right"
        elif index_tip.y < wrist.y - 0.1:  
            return "up"
        elif index_tip.y > wrist.y + 0.1:  
            return "down"

    return None  


cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

frame_skip = 5
frame_count = 0

while cap.isOpened() and not game_over:
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    if frame_count % frame_skip != 0:
        continue  

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(frame_rgb)

    gesture = None
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HandLandmark.WRIST)
            gesture = detect_gesture(hand_landmarks)

  
    if gesture in GESTURES:
        action = GESTURES[gesture]  
        print(f"🟡 Geste détecté : {gesture} -> Action : {action}")

        if gesture == "up" and acceleration_sound:
            acceleration_sound.play()

        for _ in range(3):  
            try:
                obs, reward, done, truncated, info = env.step(action)  
                env.render()

                score += reward  

               
                if score >= LEVELS[current_level]["goal"]:
                    if current_level < len(LEVELS) - 1:
                        current_level += 1
                        print(f"🚀 Niveau {current_level + 1} débloqué !")
                        setup_env(current_level)
                    else:
                        stop("🏆 VICTOIRE ! Vous avez terminé tous les niveaux !")

                if done or truncated:  
                    stop("💥 DEFAITE ! Vous avez eu un accident.")

                if np.random.random() < 0.2:  
                    overtake_count += 1
                    score += 10  

            except gym.error.ResetNeeded:
                print("🔄 Reset de l'environnement...")
                env.reset()

            time.sleep(0.1)  

   
    elapsed_time = int(time.time() - start_time)

    cv2.putText(frame, f"Niveau: {current_level + 1}", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2, cv2.LINE_AA)
    cv2.putText(frame, f"Temps: {elapsed_time}s", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(frame, f"Score: {int(score)} / {LEVELS[current_level]['goal']}", (50, 150), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    cv2.putText(frame, f"Depassements: {overtake_count}", (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2, cv2.LINE_AA)

    cv2.imshow("Jeu de Course - Détection de gestes", frame)

    # ✅ Vérifier si 'q' est pressé pour quitter
    if cv2.waitKey(1) & 0xFF == ord('q'):
        stop("🚪 JEU ARRÊTÉ PAR L’UTILISATEUR")

cap.release()
cv2.destroyAllWindows()
```
